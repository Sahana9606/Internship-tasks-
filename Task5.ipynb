```python
#Step 1 - Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
#Step 2 - Load the Datasetdf = pd.read_csv("titanic.csv")

df=sns.load_dataset("titanic")
df.head
```

```python
# Step 3 - Basic Data Exploration 

#3.1 Shape,info,Description

df.shape
df.info()
df.describe(include='all')
```

```python
#3.2 Check Missing Values

df.isnull().sum()
sns.heatmap(df.isnull(), cbar=False)
```

```python
#step 4 - Univariate Analysis

#4.1 Numerical Columns
#Histogram:Age Distribution

plt.figure(figsize=(8,5))
sns.histplot(df['age'], kde=True)
plt.title("Distribution of Age")
plt.show()
```

```python
#Histogram : Fare

plt.figure(figsize=(8,5))
sns.histplot(df['fare'], kde=True)
plt.title("Distribution of Fare")
plt.show()
```

```python
#4.2 Categorical Columns
#Passenger Class

sns.countplot(x='class', data=df)
plt.title("Passenger Class Distribution")
plt.show()
```

```python
#Survical Count

sns.countplot(x='survived', data=df)
plt.title("Survival Count")
plt.show()
```

```python
#Step 5 - bivariate Analysis 

#5.1 Gender vs survival

sns.countplot(x='sex', hue='survived', data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
#5.2 Class Vs Survival

sns.countplot(x='class', hue='survived', data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

```python
#5.3 Age Vs fare

plt.figure(figsize=(8,5))
sns.scatterplot(x='age', y='fare', hue='survived', data=df)
plt.title("Age vs Fare (Colored by Survival)")
plt.show()
```

```python
# Step 6 - Pairplot & heatmap

#6.1 Correlation Heatmap

plt.figure(figsize=(10,7))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
#6.2 Pairplot 

sns.pairplot(df[["age", "fare", "survived"]])
plt.show()
```

```python
#Step 7  Key Insights & Observations (Write in PDF)

#Univariate Insights
#Age distribution is right-skewed; most passengers are between 2040 years old.
#Majority of passengers travelled in Third Class.
#Fare has high variability with outliers on the high end.

#Bivariate Insights
#Females had a significantly higher survival rate than males.
#First Class passengers had the highest survival rate; Third Class the lowest.
#Younger passengers (kids) had relatively higher survival rates.

#Trends
#Class, sex, and fare show strong association with survival.
#Wealthier passengers (higher fare, First class) had better outcomes.



#Step 8  Summary of Findings

#Summary:
#Survival odds were strongly influenced by gender, class, and fare.
#Females and First Class passengers were prioritized during evacuation.
#Age had a moderate impactchildren had slightly better outcomes.
#Missing data existed mainly in "age" and "deck".
```
